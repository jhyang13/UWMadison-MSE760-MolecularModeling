```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load npteqfile data
npteqfile_path = r'C:/Users/jhyang/OneDrive/文档/GitHub_Projects/MSE_760/Lab3-Assignment/Problem_1-1/npteqfile.txt'
data = pd.read_csv(npteqfile_path, delim_whitespace=True, skiprows=2, 
                   names=['TimeStep', 'Temperature_K', 'Pressure_bar', 'Volume_A3', 'TotalEnergy_eV_per_atom'])
```

```python
# Normalize volume by dividing by the number of atoms (32) to get per-atom volume
data['Volume_per_atom'] = data['Volume_A3'] / 32

# Assume each timestep represents 0.001 picoseconds, convert timestep to picoseconds (ps)
data['Time_ps'] = data['TimeStep'] * 0.001

# Plot each property vs. Time (ps)
# Temperature vs Time
plt.figure(figsize=(10, 6))
plt.plot(data['Time_ps'], data['Temperature_K'], label='Temperature (K)')
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.title('Temperature vs Time')
plt.grid(True)
plt.legend()
plt.show()

# Pressure vs Time
plt.figure(figsize=(10, 6))
plt.plot(data['Time_ps'], data['Pressure_bar'], label='Pressure (bar)', color='orange')
plt.xlabel('Time (ps)')
plt.ylabel('Pressure (bar)')
plt.title('Pressure vs Time')
plt.grid(True)
plt.legend()
plt.show()

# Volume per Atom vs Time
plt.figure(figsize=(10, 6))
plt.plot(data['Time_ps'], data['Volume_per_atom'], label='Volume per Atom (Å³/atom)', color='green')
plt.xlabel('Time (ps)')
plt.ylabel('Volume per Atom (Å³/atom)')
plt.title('Volume per Atom vs Time')
plt.grid(True)
plt.legend()
plt.show()

# Total Energy per Atom vs Time
plt.figure(figsize=(10, 6))
plt.plot(data['Time_ps'], data['TotalEnergy_eV_per_atom'], label='Total Energy per Atom (eV/atom)', color='red')
plt.xlabel('Time (ps)')
plt.ylabel('Total Energy per Atom (eV/atom)')
plt.title('Total Energy per Atom vs Time')
plt.grid(True)
plt.legend()
plt.show()
```
