```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# File path to your data file, use a raw string to avoid issues with backslashes
file_path = r'C:/Users/jhyang/OneDrive/文档/GitHub_Projects/MSE_760/Lab2-Assignment/si-phonon/exercise3/reference/freq.plot'

# Load the data (q-points and corresponding frequencies)
data = np.loadtxt(file_path)

# Split the data into q-points and frequencies
q_points = data[:, 0]  # First column is q-points
frequencies = data[:, 1]  # Second column is the corresponding phonon frequencies

# Plot the phonon dispersion curve
plt.figure(figsize=(8, 6))
plt.plot(q_points, frequencies, label='Phonon Mode 1')  # Plot q-points vs frequencies

# Add labels and title to the plot
plt.xlabel('q-points')  # Label for the x-axis
plt.ylabel('Frequency (THz)')  # Label for the y-axis
plt.title('(Bulk Si Phonon Dispersion Curve')  # Title of the plot

# Display the legend and grid
#plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
